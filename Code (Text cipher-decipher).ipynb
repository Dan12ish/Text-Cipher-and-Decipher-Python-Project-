##Text Cipher and Decipher Python Project

1) Text Ciphering and deciphering (Basic Approach)

```python
def cipher(sent):
  temp=''
  for i in sent:
    temp+=(chr(ord(i)+1))
  return temp
```

```python
def decipher(sent):
  temp=''
  for i in sent:
    temp+=(chr(ord(i)-1))
  return temp
```

```python
#Testing Ciphering
sentence='Hello this is danish!'
print(cipher(sentence))
```

```python
#Testing Deciphering
sentence='Ifmmp!uijt!jt!ebojti"'
print(decipher(sentence))
```

2)Text Ciphering and Deciphering to n-places

```python
def cipher_n(sent,n):
  temp=''
  for i in sent:
    temp+=chr(ord(i)+n)
  return temp
```

```python
def decipher_n(sent,n):
  temp=''
  for i in sent:
    temp+=chr(ord(i)-n)
  return temp
```

```python
#Testing Ciphering
sentence='Hello I am danish!'
n=int(input('Enter the key to decipher: '))
print(cipher_n(sentence,n))
```

```python
#Testing Deciphering
sentence='ǻȘȟȟȢǓǼǓȔȠǓȗȔȡȜȦțǔ'
n=int(input('Enter the key to decipher: '))
print(decipher_n(sentence,n))
```

3) Text Ciphering and Deciphering with key division

```python
def cipher_key(sent,n):
  temp=''
  key=n%65
  for i in sent:
    temp+=chr(ord(i)+key)
  return temp
```

```python
def decipher_key(sent,n):
  temp=''
  key=n%65
  for i in sent:
    temp+=chr(ord(i)-key)
  return temp
```

```python
#Testing Ciphering
sent=input('Enter sentence: ')
n=int(input ('Enter key: '))
print(cipher_key(sent,n))
```

```python
#Testing Deciphering
sent=input('Enter sentence: ')
n=int(input('Enter key: '))
print(decipher_key(sent,n))
```

4) Cracking Algorithm (Brute Force)

```python
#taking a random ciphered sentenced
sent='It was a nice dinner. Lets connect some other day, probably 9PM tomorrow?'
encryption=''
for i in sent:
  encryption+=chr(ord(i)+44)

#taking the most common words. 
words=['is','the','what','are','it','me','you']

#Code to crack the encryption. 
for n in range(-65,65,1):
  guess=''
  for i in encryption:
    guess+=chr(ord(i)-n)
  for word in words:
    if (word.lower() in guess.lower()):
      print(guess)
      break
```
